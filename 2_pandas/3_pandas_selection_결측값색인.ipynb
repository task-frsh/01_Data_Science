### 2-2. loc ★★★★★

# 비교
# index [1:4] >>    1 이상 4 미만(3이하)
# loc [1:4]   >>    1 이상 4 이하
# iloc [1:4]  >>    1 이상 4 미만(3이하)

앞으로 굉장히 **자주 사용**합니다. 반드시 손에 익숙해 지도록 합시다!

```python
# loc는 ','를 기준으로 앞에 것은 행, 뒤에 것은 열
# 처음부터 끝까지 -> 모든 행, '이름'컬럼만
df.loc[:, '이름']
```

```python
# 모든 행, 열을 리스트로 지정했다는 의미는 여러 컬럼명을 넣기 위해
df.loc[:, ['이름', '생년월일']]
```

```python
# loc는 3행부터 8행이하 
# 슬라이싱이랑 다름! 주의!

df.loc[3:8, ['이름', '생년월일']]
```

```python
df.head()
```

```python
# 2행~5행, '이름' 컬럼 ~ '생년월일' 컬럼까지
df.loc[2:5, '이름':'생년월일']
```

## 3. Boolean Indexing - 조건을 활용한 색인

Boolean indexing은 **Numpy에서 배웠던 Boolean 인덱싱과 같은 원리**입니다.

```python
# 검색조건
# True, False로 되어있는 boolean indexing
# 행(Row) 출력

result = df['키'] > 180
result
```

```python
type(result)
```

```python
# 검색조건
df['키']>180

# 검색조건에 해당하는 데이터의 행만 선택
df[df['키']>180]
```

Boolean Index 로 받은 Index 를 활용해서 True인 값만 색인해 낼 수 있습니다.
* df[] 꺾쇠로 감싸주고, 그안에 Boolean Index를 넣어주시면 됩니다.

```python
df['키'] < 170
```

```python
# df[검색조건]을 이런 형태로 사용 
df[ df['키'] < 170]
```

```python
df[ df['키'] > 180 ]
```

위와 같은 방법이 매우 간편한 방법이기는 하지만, **모든 column을 출력해야만 한다는 한계**가 있습니다. 
* 특정 column과 같이 색인해 내고 싶을땐 어떻게 할까요?

```python
# df[ df['키'] > 180, '이름']
```

해결 방법 1. 맨 뒤에 출력할 column 붙히기

```python
df[ df['키'] > 180 ]['이름']
```

```python
#주의할 것은 '이름','키'를 2차원 리스트형태로 사용
df[ df['키'] > 180 ][['이름', '키']]
```

```python
# df[행 조건 필터][열 조건 필터]
```

```python
df
```

해결 방법 2. loc를 활용

```python
df.loc[ df['키'] > 180, '이름']
```

```python
df.loc[ df['키'] > 180, '이름': '성별']
```

```python
df.loc[ df['키'] > 180, ['이름', '키']]
```

저는 **방법 2를 추천** 합니다.

## 4. isin을 활용한 색인

#### **isin : 안에 있냐?**

isin을 활용한 색인은 내가 조건을 걸고자 하는 값이 

내가 정의한 list에 있을 때만 색인하려는 경우에 사용합니다.

```python
my_condition = ['플레디스', 'SM']
```

```python
# 검색조건 -> 소속사중에 플레디스 or SM이 있는지
df['소속사'].isin(my_condition)
```

```python
df.loc[ df['소속사'].isin(my_condition) ]
```

```python
df.loc[ df['소속사'].isin(my_condition),'이름' ]
```

```python
df.loc[ df['소속사'].isin(my_condition), '소속사' ]
```

* pandas 에서는 NaN => Not a Number 로 표기 된 것을 확인해 볼 수 있습니다.

```python
df
```

info() 로 NaN 값, 즉 빠진 데이터가 어디에 있는지 쉽게 요약정보로 확인할 수 있습니다.

```python
df.info()
```

### 5-2. 결측 값 다루기

Boolean 인덱싱

```python
# 데이터프레임 전체에서 결측치가 있는지 확인할때 isna()
df.isna()
```

info() 메소드를 통해서 전체적으로 어떤 column에 빠진 데이터가 있는지 알아 보았습니다.

또한, 다음과 같이 특정 column에서 빠진 값을 색출해 낼 수 있습니다.

1. Boolean 인덱싱으로 True가 return 되는 값이 NaN이라는 것을 알 수 있습니다.

```python
# series에서 null값을 확인
df['그룹'].isnull()
```

```python
# 동일한 표현
# df[df.키.isnull()]

df[df['키'].isnull()]
```

2. NaN 값만 색출해내기

```python
df['그룹'][df['그룹'].isnull()]
```

```python
df['그룹'][df['그룹'].isna()]
```

3. NaN이 아닌 값에 대하여 Boolean 인덱싱

```python
df['그룹'].notnull()
```

4. NaN이 아닌 값만 색출해내기

```python
df['그룹'][df['그룹'].notnull()]
```

```python
df.loc[ df['그룹'].notnull(), ['키', '혈액형'] ]
```
