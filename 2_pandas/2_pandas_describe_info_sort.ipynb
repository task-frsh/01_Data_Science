```python
import pandas as pd
```

## 1. 파일 읽어오기 (csv)

```python
# 작업순서

# 1. read_csv (파일명 or 경로명)
# 2. head()
# 3. tail()

# 데이터 전체 프리뷰, 맨앞-맨뒤 찍어서 전체 데이터 수량 확인
```

```python
# csv파일이 구글드라이브에 저장된 주소의 '약식 주소'
# read_csv함수는 dataFrame에 csv파일을 불러오기함
df = pd.read_csv('http://bit.ly/ds-korean-idol')
```

```python
df
```

```python
# head() : 맨 앞의 5행의 데이터를 출력한다.
# dafualt : 5라서 맨 앞의 5행 데이터를 출력한다.
# 옵션값을 변경하면 괄호 안에 인수 지정
df.head()
```

```python
# tail() : 맨 끝에서5개의 데이터를 출력한다.
df.tail()
```

## 7. 정렬하기 

### 7-1. 오름차순 정렬 (default)

```python
# sort_index()함수는 index를 기준으로 정렬해주는 함수
# dafault : 오름차순

df.sort_index()
```

### 7-2. 내림차순 index 정렬

```python
# ascending=True : 오름차순
# ascending=False : 내림차순

df.sort_index(ascending=False)
```

## 8. column 별로 정렬

### 8-1. 오름차순 정렬

```python
# dafualt : NaN값은 제일 마지막에 몰아둠
# NaN값은 지워버리거나, 채운다

df.sort_values(by='키')
```

### 8-2. 내림차순 정렬

```python
# ascending=False == descending

df.sort_values(by='키', ascending=False)
```

### [BONUS] 복수 정렬

```python
# 키로 정렬했는데 똑같다?? 그러면 2번 기준, 3번기준으로 정렬함
# 정렬 기준 1번 :'키', 키가 동일한 경우 정렬기준 2번

df.sort_values(by=['키', '브랜드평판지수'])

# 리스트로 표현
```

```python
df.sort_values(by=['키', '브랜드평판지수'], ascending=False)
```
