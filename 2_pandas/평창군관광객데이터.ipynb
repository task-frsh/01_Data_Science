```python
df.columns

# 정리
# 1> 2010년 '1월' ~ '12월'까지 관광지 1곳에 다녀간 모든 사람수 계산
# 2> 2010년 '1월' ~ '12월'까지 모든 관광지에 다녀간 사람 수 합산
# 3> 연도별로 비교해서 가장 방문객이 많았던 해 찾기



# 1> 2010년 '1월' ~ '12월'까지 관광지 1곳에 다녀간 모든 사람수 계산


# df라는 DataFrame에서 index 0번째 행만 출력해줘
# print(df.loc[0])

# df에서 index 0번째 행 '1월'~'12월'까지 출력해줘
print(df.loc[0, '1월':'12월'])

# df에서 index 0번째 행 '1월'~'12월'값을 더해줘

df['총방문객수'] = df.loc[0, '1월':'12월'].sum(axis=0)
df['총방문객수']

#### 꼭 검증하기 ####
```

```python
# 2> 2010년 '1월' ~ '12월'까지 모든 관광지에 다녀간 사람 수 합산

# 2010년 데이터만 골라내기

df_2010 = df.loc[df['연도'] == 2010]
df_2010



# 모든 관광지의 2010.1~12월 방문객수를 더해
# 총 방문객수 계산

df_2010['총방문객수'] = df_2010.loc[:, '1월':'12월'].sum(axis=1)

print(df_2010['총방문객수'])

print("*" * 50)

# 2010년 방문객 수 합산

print(df_2010['총방문객수'].sum())
```

```python
# 3> 연도별로 비교해서 가장 방문객이 많았던 해 찾기


# 결측치 전처리
# 결측치 있는 행 : 2022.4~12

# df에서 결측치가 하나라도 있을 모두 삭제할거야
# 변경된 결과는 원본 df에 반영해줘

df.dropna(how='any', inplace=True)
df
```

```python

# 2010~2022까지 순서대로 연도별 관광지별 관광객 수 합산


def yearTotal(year) :
    
    # 특정 해의 데이터만 선택해서 데이터 프레임 만들기
    df_year = df.loc[df['연도'] == year]

    # 각 연도의 1~12월까지 모든 관광지에 방문한 총 방문객 수를 계산해
    df_year['총방문객수'] = df_year.loc[:, '1월':'12월'].sum(axis=1)

    # 전체 방문객 수 합산해
    total = int(df_year['총방문객수'].sum())
    return total


for year in range(2010, 2022) :
    print(f'{year}년 : {yearTotal(year)}')
```

[같은 데이터를 도돌이표처럼 계속 쓰니까 시스템이 불편해함]
-> copy해서 쓰는 걸로 수정



```python

# 2010~2022까지 순서대로 연도별 관광지별 관광객 수 합산


def yearTotal(year) :
    
    # 특정 해의 데이터만 선택해서 데이터 프레임 만들기
    # copy해서 다른 메모리에 df_year를 저장함
    df_year = df.loc[df['연도'] == year].copy()

    # 각 연도의 1~12월까지 모든 관광지에 방문한 총 방문객 수를 계산해
    df_year['총방문객수'] = df_year.loc[:, '1월':'12월'].sum(axis=1)

    # 전체 방문객 수 합산해
    total = int(df_year['총방문객수'].sum())
    return total


for year in range(2010, 2022) :
    print(f'{year}년 : {yearTotal(year)}')
```
