```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] =False
```

```python
import pandas as pd

# 엑셀 가져오기
FilePath = '/content/drive/MyDrive/Colab Notebooks/Data Science/4_데이터분석예제/data/06002 여가시간_202403.xlsx'

df = pd.read_excel(FilePath, header=6)
```

```python
df.head()
```

```python
df.tail(10)
```

```python
# 맨 아래쪽 필요없는 행 지우기 (~35번까지만 남기기)

df = df.iloc[:36].reset_index(drop=True)
```

```python
df.tail()
```

```python
# '-' 값을 0으로 대체
df['70대 이상'] = df['70대 이상'].replace('-', 0).fillna(0)

# 데이터 타입을 int로 변경
df['70대 이상'] = df['70대 이상'].astype(float)

# 소수점 1자리까지
df = df.round(1)
```

```python
# 필요없는 Unnamed: 0 열 삭제

df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
# 컬럼명 변경

df.columns = ['년도', '분류', '전체', '남자', '여자', '10대', '20대',
       '30대', '40대', '50대', '60대', '70대 이상']
```

```python
df.head()
```

```python
# 분류 - 요일평균 여가시간 빈 부분 채우기

df['분류'] = df['분류'].fillna('요일평균 여가시간')
df
```

```python
# 연도 정보가 NaN인 부분을 적절히 채우기 위해 fill 사용

df['년도'] = df['년도'].ffill()
df



# 예전 함수인데.. 이거 이제 안됨 > df['년도'] = df['년도'].fillna(method='ffill')
```

```python
# 요일평균 여가시간 데이터프레임 추출
df_weekly = df[df['분류'] == '요일평균 여가시간'].reset_index(drop=True)

# 평일 여가시간 데이터프레임 추출
df_weekday = df[df['분류'] == '평일 여가시간'].reset_index(drop=True)

# 휴일 여가시간 데이터프레임 추출
df_weekend = df[df['분류'] == '휴일 여가시간'].reset_index(drop=True)
```

```python
df_weekly.drop(columns=['분류'],inplace=True)
df_weekday.drop(columns=['분류'],inplace=True)
df_weekend.drop(columns=['분류'],inplace=True)
```

```python
df_weekly.set_index('년도',inplace=True)
df_weekday.set_index('년도',inplace=True)
df_weekend.set_index('년도',inplace=True)
```

```python
# 각 데이터프레임의 결과 확인
print("요일평균 여가시간 데이터프레임:")
print(df_weekly)

print("\n평일 여가시간 데이터프레임:")
print(df_weekday)

print("\n휴일 여가시간 데이터프레임:")
print(df_weekend)
```

```python
# 막대 그래프 그리기 함수 정의
def plot_bar(df, title):
    df.plot(kind='bar', figsize=(10, 6))
    plt.title(title)
    plt.xlabel('년도')
    plt.ylabel('시간(시간/일)')
    plt.legend(title='연령대', bbox_to_anchor=(1, 1)) # bbox_to_anchor=(1, 1) ,loc = 'best' # 범례 박스 왼쪽 하단 코너 좌표
    plt.tight_layout()
    plt.show()

# 각 데이터프레임의 막대 그래프 그리기
plot_bar(df_weekly, '요일평균 여가시간')
plot_bar(df_weekday, '평일 여가시간')
plot_bar(df_weekend, '휴일 여가시간')
```
