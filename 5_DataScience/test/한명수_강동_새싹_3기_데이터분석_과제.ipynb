````python
import pandas as pd

# 고정된 학생 데이터 생성

students = ['학생1', '학생2', '학생3', '학생4', '학생5']
subjects = ['국어', '영어', '수학', '과학', '사회']

data = {
    '국어' : [85, 78, 75, 80, 95],
    '영어' : [88, 92, 78, 85, 84],
    '수학' : [100, 67, 78, 82, 76],
    '과학' : [82, 95, 67, 89, 84],
    '사회' : [79, 85, 87, 90, 93],
    '학생' : students
}

df = pd.DataFrame(data)
````

## 3. 과목의 총점과 평균 컬럼 추가

image.png

````python
#df를 복제해서 df1을 만들거야.

df1 = df.copy()
df1
````

````python
df1['평균점수'] = df1.mean(axis=1)
````

````python
df1.head()
````

````python
df1['총점'] = df1.sum(axis=1) - df1['평균점수']
````

````python
df1
````

````python
df1.info()
````

````python
# 컬럼명 순서 변경
# 국어	영어	수학	과학	사회	총점 평균점수

df1 = df1[['국어', '영어', '수학', '과학', '사회', '총점', '평균점수']]
````

````python
# df1['총점'] 속성을 int64로 형변환한다

df1['총점'] = df1['총점'].astype('int64')
````
