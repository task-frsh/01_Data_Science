## 파일 불러오기

### 서울등록인구 popSeoul

```python
# popSeoul

csvPath = '/content/drive/MyDrive/Colab Notebooks/Data Science/4_데이터분석예제/data/등록인구_20240807134714.csv'
```

```python
import pandas as pd
import numpy as np

popSeoul = pd.read_csv(csvPath, encoding='utf-8')
```

```python
popSeoul.head()
```

```python
popSeoul.loc[0]
```

```python
popSeoul.columns = popSeoul.loc[0]
popSeoul.head()
```

```python
# popSeoul 0번행 삭제

popSeoul = popSeoul.drop(0)
popSeoul.head()
```

```python
popSeoul.columns
```

```python
# 불필요한 컬럼 제외
# iloc 이용해서 필요한 컬럼만 골라내기
# [1,3,6,9,13]

popSeoul = popSeoul.iloc[:, [1,3,6,9,13]]
popSeoul.head()
```

```python
# 컬럼명 변경

popSeoul.columns = ['구별', '인구', '내국인', '외국인', '고령자']
popSeoul.head()
```

```python
# 필요없는 행 (1, 2)행 삭제

popSeoul = popSeoul.drop([1, 2])
popSeoul.head()
```

```python
# 인덱스 값 초기화

popSeoul = popSeoul.reset_index(drop=True)
popSeoul.head()
```

```python
popSeoul.info()
popSeoul.tail()
```

### CCTV 연도별

```python
# cctv 연도별 데이터. 필요한 셀만 가져온다

xlsxPath1 = '/content/drive/MyDrive/Colab Notebooks/Data Science/4_데이터분석예제/data/서울시 자치구 (연도별) CCTV 설치현황_240630.xlsx'

cctvSeoul = pd.read_excel(xlsxPath1, header=2, usecols='C,D,E,F,G,H,I,J,K,L,M,N')
cctvSeoul.head()
```

```python
# cctvSeoul 0번행 삭제

cctvSeoul = cctvSeoul.drop(0)
cctvSeoul.head()
```

```python
# index 초기화

cctvSeoul = cctvSeoul.reset_index(drop=True)
cctvSeoul.head()
```

```python
cctvSeoul.columns = ['구별', 'CCTV대수', '2016년이전CCTV ', '2016년', '2017년', '2018년', '2019년',
       '2020년', '2021년', '2022년', '2023년', '2024년2분기']

cctvSeoul.head()
```

## 파일 합치기

```python
# popSeoul, cctvSeoul 데이터프레임 합치기

result = pd.merge(popSeoul, cctvSeoul, on='구별')
result.head()
```
