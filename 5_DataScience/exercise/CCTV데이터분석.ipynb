```python
# df.columns[0] 컬럼 삭제
pop_df.drop(pop_df.columns[0], axis=1, inplace=True)
pop_df.head()
```

```python
columnName = []
pop_df.iloc[0][0]="구별"
pop_df.iloc[1][0]=""

for name in pop_df.iloc[0]:
    columnName.append(name.split()[0])

for i, name in enumerate(pop_df.iloc[1]):
    columnName[i] += ' '+name

pop_df.columns = columnName
pop_df.head()
```

```python
#df의 0,1,2행 삭제
pop_df.drop(pop_df.index[[0,1,2]], inplace=True)
pop_df.head()

pop_df.head()
```

```python
CCTVFile = fileNames[3]

# xlsx파일 불러오기
import pandas as pd

cctv_df = pd.read_excel(os.path.join(csvPath, CCTVFile))

# 1, 4~7번째 컬럼 삭제
cctv_df.drop(cctv_df.columns[[0,4,5,6,7]], axis=1, inplace=True)
cctv_df.columns = cctv_df.iloc[1]
# 0,1, 2행 삭제
cctv_df.drop(cctv_df.index[[0,1,2]], inplace=True)
cctv_df.reset_index(drop=True, inplace=True)
cctv_df.head()
```
