## 2. ndarray 생성하기

```
# 파이썬의 리스트는 다양한 데이터 타입을 포함할 수 있다.
# numpy에서는 ndarray 생성시 데이터 타입을 하나로 통일하게 된다.
arr = np.array([1., 2, 3, 4], dtype=int)
```

```
arr
```

```
type(arr)
```

```
arr0 = np.array([1.,2.,3.,4.])
print(arr0)
type(arr0)
```
